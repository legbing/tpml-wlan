```python
for i in v1:
    file = path + i + ".txt"
    f = open(file, "r")

    line = f.readlines()
    line = [x.strip() for x in line]
    for count in range(0, 17*100, 17):
            throughput.append(line[1 + count])
            airtime.append(line[2 + count])
            RSSI.append(line[3 + count])
            interference_map.append(line[4 + count:16+count])
            SINR.append(line[16 + count])
            
for i in v2:
    file = path + i + ".txt"
    f = open(file, "r")

    line = f.readlines()
    line = [x.strip() for x in line]
    for count in range(0, 13*100, 13):
            throughput.append(line[1 + count])
            airtime.append(line[2 + count])
            RSSI.append(line[3 + count])
            interference_map.append(line[4 + count:12+count])
            SINR.append(line[12 + count])
```
