```python
for i in range(1,5):
    file = path + str(i) + ".txt"
    f = open(file, "r")

    line = f.readlines()
    line = [x.strip() for x in line]
    for count in range(0, 9*50 + 100*(i-1), 7+2*i):
            throughput.append(line[1 + count])
            airtime.append(line[2 + count])
            RSSI.append(line[3 + count])
            interference_map.append(line[4 + count:(4+count)+2*i + 2])
            SINR.append(line[4+ 2*i + 2 + count])
```
